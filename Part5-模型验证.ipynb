```python
from onnxInference import *
```

## 模型推理配置参数

```python
# 模型路径
model_path = "./models/dy_model_new.onnx"
# 推理数据路径
path_inference = "./dataset/推理输出/推理数据.csv"  
# 文件保存路径
save_path = "./dataset/推理输出/推理数据_修改后结果.csv"
# 标签类别映射文件名路径
config_name = "./config/标签类别映射/config72.json"
```

## 单条推理

```python
texts = ["人行前置机A到人行2900A异常，172.18.81.37 BPH001072 人行前置机B到人行2900A异常"]  # 添加更多文本样本
label_list, pred_probs = main_inferance(texts, config_name, model_path)
print(label_list)
print(pred_probs)
# 标签和概率一一对应
```

## 批量推理

```python
# 批次大小
batch_size = 200
inference_df = batch_inference(path_inference, batch_size, config_name, model_path)
```

## 文件保存

```python
inference_df.to_csv(save_path, index=False)
```
